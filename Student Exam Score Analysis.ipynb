```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Expanded_data_with_more_features[1].csv")
```

```python
#droping unnamed column
dp=df.drop("Unnamed: 0",axis=1)
print(df.head())
```

```python
#gender distribution
plt.figure(figsize=(4,4))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb,annot=True,cmap='binary')
plt.title("Relationship between Parent's Education and Students Score")
plt.show()
```

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb1,annot=True)
plt.title("Relationship between Parent's Maritial status and Students Score")
plt.show()
```

```python
gb2=df.groupby("IsFirstChild").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb2)
plt.figure(figsize=(4,4))
sns.heatmap(gb2,annot=True,cmap='Greens')
plt.title("Relationship if student is first child and Students Score")
plt.show()
```

```python
gb3=df.groupby("TransportMeans").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb3)
plt.figure(figsize=(4,4))
sns.heatmap(gb3,annot=True,cmap='PiYG')
plt.title("Relationship between TransportMeans and Students Score")
plt.show()
```

```python
gb4=df.groupby("PracticeSport").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb4)
plt.figure(figsize=(4,4))
sns.heatmap(gb4,annot=True)
plt.title("Relationship between Practising sports and Students Score")
plt.show()
```

```python
groupA=df.loc[(df['EthnicGroup']=="group A")].count()
groupB=df.loc[(df['EthnicGroup']=="group B")].count()
groupC=df.loc[(df['EthnicGroup']=="group C")].count()
groupD=df.loc[(df['EthnicGroup']=="group D")].count()
groupE=df.loc[(df['EthnicGroup']=="group E")].count()

label=["group A","group B","group C","group D","group E",]
mylist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mylist,labels=label,autopct="%1.2f%%",colors=sns.color_palette('Set2'))
plt.title("Distribution of Ethnic Groups")
print(mylist)

plt.show()
```
